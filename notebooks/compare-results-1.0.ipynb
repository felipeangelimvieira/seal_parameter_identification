```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

with open("results/frequency_eiv_results.pickle", "rb") as f:
    eiv_results = pickle.load(f)

with open("results/frequency_jio_results.pickle", "rb") as f:
    jio_results = pickle.load(f)

with open("results/optimization_sin_results.pickle", "rb") as f:
    opt_results = pickle.load(f)
    
    
Cs_eiv = np.array([d["C"] for d in eiv_results])
Ks_eiv = np.array([d["K"] for d in eiv_results])
Fs_eiv = np.array([d["freq"] for d in eiv_results])

Cs_jio = np.array([d["C"] for d in jio_results])
Ks_jio = np.array([d["K"] for d in jio_results])
Fs_jio = np.array([d["freq"] for d in jio_results])

Cs_opt = np.array([d["C"] for d in opt_results])
Ks_opt = np.array([d["K"] for d in opt_results])
Fs_opt = np.array([d["freq"] for d in opt_results])
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,10))
axs[0,0].plot(Fs_eiv, Ks_eiv[:, 0, 0])
axs[0,0].plot(Fs_jio, Ks_jio[:, 0, 0])
axs[0,0].scatter(Fs_opt, Ks_opt[:, 0, 0], s=21, color="green")
axs[0,0].set_title("Kxx")

axs[0,1].plot(Fs_eiv, Ks_eiv[:, 0, 1])
axs[0,1].plot(Fs_jio, Ks_jio[:, 0, 1])
axs[0,1].scatter(Fs_opt, Ks_opt[:, 0, 1], s=21, color="green")
axs[0,1].set_title("Kxy")

axs[1,0].plot(Fs_eiv, Ks_eiv[:, 1, 0])
axs[1,0].plot(Fs_jio, Ks_jio[:, 1, 0])
axs[1,0].scatter(Fs_opt, Ks_opt[:, 1, 0], s=21, color="green")
axs[1,0].set_title("Kyx")

axs[1,1].plot(Fs_eiv, Ks_eiv[:, 1,1])
axs[1,1].plot(Fs_jio, Ks_jio[:, 1,1])
axs[1,1].scatter(Fs_opt, Ks_opt[:, 1, 1], s=21, color="green")
axs[1,1].set_title("Kyy")


fig, axs = plt.subplots(2,2, figsize=(15,10))
axs[0,0].plot(Fs_eiv, Cs_eiv[:, 0, 0])
axs[0,0].plot(Fs_jio, Cs_jio[:, 0,0])
axs[0,0].scatter(Fs_opt, Cs_opt[:, 0, 0], s=21, color="green")
axs[0,0].set_title("Cxx")

axs[0,1].plot(Fs_eiv, Cs_eiv[:, 0, 1])
axs[0,1].plot(Fs_jio, Cs_jio[:, 0, 1])
axs[0,1].scatter(Fs_opt, Cs_opt[:, 0, 1], s=21, color="green")
axs[0,1].set_title("Cxy")

axs[1,0].plot(Fs_eiv, Cs_eiv[:, 1, 0])
axs[1,0].plot(Fs_jio, Cs_jio[:, 1, 0])
axs[1,0].scatter(Fs_opt, Cs_opt[:, 1, 0], s=21, color="green")
axs[1,0].set_title("Cyx")

axs[1,1].plot(Fs_eiv, Cs_eiv[:, 1, 1])
axs[1,1].plot(Fs_jio, Cs_jio[:, 1, 1])
axs[1,1].scatter(Fs_opt, Cs_opt[:, 1, 1], s=21, color="green")
axs[1,1].set_title("Cyy")

fig.show()
```
